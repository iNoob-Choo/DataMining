```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def linear(u,m,c):
    #y=mu+c
    return m*u+c


v=linear(u,2,10)
plt.plot(u,v)
```

```python
def step(u):
    v=np.zeros_like(u)
    for i,ui in enumerate(u):
        if ui>0:
            v[i]=1
        else:
            v[i]=0
    return v

x=step(u)
plt.plot(u,x)

```

```python
def step_refine(u):
    return (u>0)*1

y=step_refine(u)
plt.plot(u,y)

   
```

```python
def tanh(u):
    return np.tanh(u)
v=tanh(u)

plt.plot(u,v)
```

```python
def sigmoid(u):
    return 1/(1+np.exp(-u))

v=sigmoid(u)
plt.plot(u,v)
```

```python
def Relu(u):
   
    return (u>0)*u

v=Relu(u)

plt.plot(u,v)

    
```
